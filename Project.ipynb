```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
def fx(y,z):
  return  - 0.64+(1+z)*y 
def fy(vy):
  return vy
def gy(y,z):
  return - (1+z)**2*y+0.64*(1+z)
  
def fz(vz):
  return vz
def gz(y,z):
  return 0.64*y-(1+z)*y**2


def RK_4th_Order(x0,y0,z0,vy0,vz0):
  """This Function is used to solve 1st and 2nd order ordinary differential equation in Rk-method """
  """return value :x,y,z"""
  h = 0.1
  t0 = 0
  x = []
  y = []
  z = []
  t = []
  for i in range(1,3000):
    # for solution of x
    k1x = h*fx(y0,z0)
    k2x = h*fx(y0 + k1x/2, z0 + k1x/2)
    k3x = h*fx(y0 + k2x/2, z0 + k2x/2)
    k4x = h*fx(y0 + k3x, z0 + k3x)

    # for solution of y
    k1y = h*fy(vy0)
    l1y = h*gy(y0,z0)
    k2y = h*fy(vy0 + l1y/2)
    l2y = h*gy(y0 + k1y/2, z0 + k1y/2)
    k3y = h*fy(vy0 + l2y/2)
    l3y = h*gy(y0 + k2y/2, z0 + k2y/2)
    k4y = h*fy(vy0 + l3y)
    l4y = h*gy(y0 + k3y, z0 + k3y)

    # for solution of z
    k1z = h*fz(vz0)
    l1z = h*gz(y0,z0)
    k2z = h*fz(vz0 + l1z/2)
    l2z = h*gz(y0 + k1z/2, z0 + l1z/2)
    k3z = h*fz(vz0 + l2z/2)
    l3z = h*gz(y0 + k2z/2, z0 + l2z/2)
    k4z = h*fz(vz0 + l3z)
    l4z = h*gz(y0 + k3z, z0 + l3z)

    xi = x0 + 1/6 *(k1x + 2*k2x + 2*k3x + k4x)
    yi = y0 + 1/6 *(k1y + 2*k2y + 2*k3y + k4y)
    zi = z0 + 1/6 *(k1z + 2*k2z + 2*k3z + k4z)
    vyi = vy0 + 1/6 *(l1y + 2*l2y + 2*l3y + l4y)
    vzi = vz0 + 1/6 *(l1z + 2*l2z + 2*l3z + l4z)
    x.append(x0)
    y.append(y0)
    z.append(z0)
    t.append(t0)
    t0 += 0.1
    x0 = xi
    y0 = yi
    z0 = zi
    vy0 = vyi
    vz0 = vzi 

  return x,y,z,t
```

```python
x,y,z,t = RK_4th_Order(0,0.66,0,0,0.08)
```
